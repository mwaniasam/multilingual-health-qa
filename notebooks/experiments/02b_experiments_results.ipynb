# 02 Experiments and Results

This notebook runs the 8 experiments and displays the final comparison table. The primary metric is Hamming Loss.

```python
import os
import subprocess
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
venv_python = ROOT / '.venv' / 'bin' / 'python'
script = ROOT / 'scripts' / 'run_experiments.py'
runner = venv_python if venv_python.exists() else Path(sys.executable)

env = os.environ.copy()
if venv_python.exists():
    env['MPLCONFIGDIR'] = str(ROOT / 'cache' / 'matplotlib')

completed = subprocess.run(
    [str(runner), str(script)],
    cwd=ROOT,
    check=True,
    text=True,
    capture_output=True,
    env=env,
)
print(completed.stdout, end='')
if completed.stderr:
    print(completed.stderr, end='')
```
